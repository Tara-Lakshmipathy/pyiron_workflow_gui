````python
# After 20 epochs
```python
def split_string(string):
    substrings = []
    delimiters = []
    in_quotes = False
    for char in string:
        if char in ('\'', '\"'):
            in_quotes = not in_quotes
        elif in_quotes:
            continue
        elif char.isalnum():
            continue
        else:
            delimiters.append(char)
    for delimiter in delimiters:
        substrings.append('')
    for char in string:
        if char.isalnum():
            continue
        elif in_quotes:
            continue
        else:
            index = delimiters.index(char)
            substrings[index] += char
    return substrings
```
````

````python
# Desired
#```python
def compare_arrays(arrays):
    common_elements = {}
    
    for array in arrays:
        unique_elements = set(array)
        
        for element in unique_elements:
            if element in common_elements:
                common_elements[element] += 1
            else:
                common_elements[element] = 1
    
    result = []
    for element, count in common_elements.items():
        if count == len(arrays):
            result.append((element, count))
    
    return result
````

````python
# After 20 epochs
#```python
def find_common_elements(arrays):
    common_elements = set()
    common_count = {}
    
    for array in arrays:
        unique_elements = set(array)
        
        common_elements.update(unique_elements)
        
        for element in common_elements:
            if element in common_count:
                common_count[element] += 1
            else:
                common_count[element] = 1
    
    common_elements_sorted = sorted(common_elements, key=lambda x: common_count[x], reverse=True)
    
    return common_elements_sorted, common_count
````
